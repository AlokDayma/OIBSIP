```python
sns.histplot(x = 'Estimate_Employed',hue='Region_D',data=df)
plt.title('India Unemployment')
plt.show()
```

```python
sns.histplot(x = 'Rate(%)',y= hue='Region_D',data=df)
plt.title('India Unemployment')
plt.show()
```

```python
import plotly.express as px
unemployment = df[["Region", "Region_D", "Rate(%)"]]
figure = px.sunburst(unemployment, path=["Region_D", "Region"], 
                     values="Rate(%)", 
                     width=700, height=700, color_continuous_scale="RdY1Gn", 
                     title="Unemployment Rate in India")
figure.show()
```
